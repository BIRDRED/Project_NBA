```python
NBA_Seasons = NBA_Seasons.drop(['Age','Rk','Lg','Ht','Wt','G','MP','TRB','FT/FGA','ORtg','Pace', 'FG%', '3P%', 'FT%', 'eFG%',
       'TOV%', 'ORB%','3PA','FGA','FTA'],axis=1)
```

```python
datashape = NBA_Seasons.columns.shape[0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("ticks")

fig, ax = plt.subplots(1, 1, figsize=(24, 10))
NBA_Seasons.plot(ax=ax)
sns.despine(offset=20, trim=True)
```

```python
ccf_matrix = pd.DataFrame(
    np.zeros((datashape,datashape)),
    index=NBA_Seasons.columns,
    columns=NBA_Seasons.columns)
ccf_matrix
```

```python
for col1 in NBA_Seasons.columns:
    for col2 in NBA_Seasons.columns:
        result = ccf(NBA_Seasons[col1], NBA_Seasons[col2], unbiased=False)
        ccf_matrix.loc[col1, col2] = 1. - max(result) 
        ccf_matrix.loc[col2, col1] = 1. - max(result)
ccf_matrix
```

```python
sns.clustermap(ccf_matrix, figsize=(14, 14), method="ward", cmap="coolwarm_r")
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

sns.set_style("darkgrid")
fig, ax = plt.subplots(1, 1, figsize=(16, 8))

linkage_matrix = linkage(ccf_matrix, method="average")
dendrogram = dendrogram(linkage_matrix)
ax.set_xticklabels(ccf_matrix.columns[dendrogram["leaves"]]);
```

```python
clusters = fcluster(linkage_matrix, 3, criterion="maxclust")
```

```python
from collections import defaultdict
groups = defaultdict(list)
for idx, clust in enumerate(clusters):
    groups[clust].append(ccf_matrix.columns[idx])
```

```python
fig, axes = plt.subplots(max(clusters), 1, figsize=(20, 14), sharex=True)

sns.set_context("talk")
sns.set_style("ticks")


for idx, symbol_cluster in groups.items():
    NBA_Seasons[symbol_cluster].plot(ax=axes[idx - 1])
fig.tight_layout()
```
